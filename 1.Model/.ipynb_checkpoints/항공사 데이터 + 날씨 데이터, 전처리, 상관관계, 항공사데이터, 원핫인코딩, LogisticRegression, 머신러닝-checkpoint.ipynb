```python
import pandas as pd
import numpy as np
```

```python
csv_path = "flight_weather_merged.csv"
df = pd.read_csv(csv_path)

print("âœ… CSV ë¡œë“œ ì™„ë£Œ")
print("í–‰ ìˆ˜:", len(df))
print("ì»¬ëŸ¼ ìˆ˜:", df.shape[1])
df.head()
```

```python
# =====================================================
# 2ï¸âƒ£ íƒ€ê²Ÿ ë³€ìˆ˜ ìƒì„± (ì§€ì—° ì—¬ë¶€)
# =====================================================
def make_delay_label(x):
    if pd.isna(x):
        return 0
    if any(k in x for k in ["ì§€ì—°", "ê²°í•­", "íšŒí•­"]):
        return 1
    return 0

df["is_delay"] = df["RMK_KOR"].apply(make_delay_label)
print("\nğŸ“Œ is_delay ë¶„í¬")
print(df["is_delay"].value_counts())
```

```python
# =====================================================
# 3ï¸âƒ£ ë‚ ì§œ / ì‹œê°„ íŒŒìƒ (ê¸°ì¡´ ë°©ì‹ ìœ ì§€)
# =====================================================
df["flight_datetime"] = pd.to_datetime(df["flight_datetime"])

df["flight_date"] = df["flight_datetime"].dt.date
df["weekday"] = df["flight_datetime"].dt.weekday
df["month"] = df["flight_datetime"].dt.month
df["hour"] = df["flight_datetime"].dt.hour
```

```python
# =====================================================
# 4ï¸âƒ£ ëª¨ë¸ ì…ë ¥ì—ì„œ ì œì™¸í•  ì»¬ëŸ¼ ì •ì˜
# =====================================================
exclude_cols = [
    # íƒ€ê²Ÿ
    "is_delay",

    # ì›ë³¸ ë‚ ì§œ/ì‹œê°„
    "FLIGHT_DATE",
    "flight_date",
    "STD",
    "ETD",
    "flight_datetime",
    "weather_datetime",
    "ì¼ì‹œ",

    # í…ìŠ¤íŠ¸ ì„¤ëª… (ëˆ„ìˆ˜ ìœ„í—˜)
    "RMK_KOR",

    # ì¤‘ë³µ / ì˜ë¯¸ ì—†ëŠ” ì‹ë³„ì
    "ì§€ì ëª…"
]

exclude_cols = [c for c in exclude_cols if c in df.columns]
```

```python
# =====================================================
# 5ï¸âƒ£ ë²”ì£¼í˜• / ìˆ˜ì¹˜í˜• ì»¬ëŸ¼ (ğŸ”¥ ê°œì„ ëœ ëª…ì‹œì  ì„ íƒ)
# =====================================================

# âœ… ë²”ì£¼í˜• (flight í•µì‹¬ + weather í•µì‹¬)
cat_cols = [
    # flight
    "AIRLINE_KOREAN",
    "BOARDING_KOR",
    "ARRIVED_KOR",
    "LINE",
    "IO",

    # weather (ë²”ì£¼í˜•)
    "ì¼ê¸°í˜„ìƒ",
    "1ì¸µ ìš´í˜•"
]

cat_cols = [c for c in cat_cols if c in df.columns]

# âœ… ìˆ˜ì¹˜í˜• (ì‹œê°„ íŒŒìƒ + ë‚ ì”¨ ìˆ˜ì¹˜í˜•)
numeric_candidates = df.select_dtypes(include=["int64", "float64"]).columns.tolist()

# ìˆ«ìì²˜ëŸ¼ ìƒê¸´ ID / ì½”ë“œ ì œê±°
exclude_numeric_like_id = [
    "ì§€ì ",
    "í˜„ìƒë²ˆí˜¸(êµ­ë‚´ì‹)"
]

num_cols = [
    c for c in numeric_candidates
    if c not in exclude_numeric_like_id
    and c not in exclude_cols
]

print("\nğŸ“Š ìµœì¢… ìˆ˜ì¹˜í˜• ì»¬ëŸ¼ ìˆ˜:", len(num_cols))
print(num_cols)

print("\nğŸ· ìµœì¢… ë²”ì£¼í˜• ì»¬ëŸ¼ ìˆ˜:", len(cat_cols))
print(cat_cols)
```

```python
# =====================================================
# 6ï¸âƒ£ ê²°ì¸¡ì¹˜ ì²˜ë¦¬
# =====================================================

# ìˆ˜ì¹˜í˜• â†’ ì¤‘ì•™ê°’
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# ë²”ì£¼í˜• â†’ UNKNOWN
for col in cat_cols:
    df[col] = df[col].fillna("UNKNOWN")
```

```python
# =====================================================
# 7ï¸âƒ£ Train / Test ë¶„ë¦¬ (ë‚ ì§œ ê¸°ì¤€)
# =====================================================
df = df.sort_values("flight_date")

split_date = df["flight_date"].quantile(0.8)

train_df = df[df["flight_date"] <= split_date]
test_df  = df[df["flight_date"] > split_date]

print("\nğŸ“¦ ë°ì´í„° ë¶„ë¦¬")
print("Train:", len(train_df), "Test:", len(test_df))

X_cols = num_cols + cat_cols

X_train = train_df[X_cols]
y_train = train_df["is_delay"]

X_test  = test_df[X_cols]
y_test  = test_df["is_delay"]
```

```python
# =====================================================
# ğŸ” Aï¸âƒ£ ìˆ˜ì¹˜í˜• ë³€ìˆ˜ â†” is_delay ìƒê´€ê´€ê³„
# =====================================================
from scipy.stats import pointbiserialr

num_corr_results = []

for col in num_cols:
    try:
        corr, pval = pointbiserialr(train_df[col], train_df["is_delay"])
        num_corr_results.append({
            "feature": col,
            "corr": corr,
            "abs_corr": abs(corr),
            "p_value": pval
        })
    except Exception as e:
        print(f"âš ï¸ {col} ê³„ì‚° ì‹¤íŒ¨:", e)

num_corr_df = (
    pd.DataFrame(num_corr_results)
    .sort_values("abs_corr", ascending=False)
)

print("\nğŸ“Š ìˆ˜ì¹˜í˜• ë³€ìˆ˜ ìƒê´€ê´€ê³„ (ìƒìœ„ 20ê°œ)")
display(num_corr_df.head(20))
```

```python
# =====================================================
# ğŸ” Bï¸âƒ£ ë²”ì£¼í˜• ë³€ìˆ˜ â†” is_delay (Chi-square)
# =====================================================
from scipy.stats import chi2_contingency

cat_corr_results = []

for col in cat_cols:
    contingency = pd.crosstab(train_df[col], train_df["is_delay"])

    # ëª¨ë“  ê°’ì´ í•œìª½ í´ë˜ìŠ¤ë§Œ ê°€ì§€ë©´ ìŠ¤í‚µ
    if contingency.shape[1] < 2:
        continue

    chi2, p, _, _ = chi2_contingency(contingency)

    cat_corr_results.append({
        "feature": col,
        "chi2": chi2,
        "p_value": p
    })

cat_corr_df = (
    pd.DataFrame(cat_corr_results)
    .sort_values("chi2", ascending=False)
)

print("\nğŸ· ë²”ì£¼í˜• ë³€ìˆ˜ Chi-square ê²°ê³¼")
display(cat_corr_df)
```

```python
# =====================================================
# 8ï¸âƒ£ ì „ì²˜ë¦¬ + ëª¨ë¸ ì •ì˜
# =====================================================
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

class_weight = {0: 1, 1: 70}

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
        ("num", "passthrough", num_cols)
    ]
)

model = Pipeline([
    ("prep", preprocessor),
    ("clf", LogisticRegression(
        max_iter=5000,
        class_weight=class_weight,
        n_jobs=-1
    ))
])
```

```python
# =====================================================
# 9ï¸âƒ£ ëª¨ë¸ í•™ìŠµ
# =====================================================
model.fit(X_train, y_train)
print("\nâœ… ëª¨ë¸ í•™ìŠµ ì™„ë£Œ")
```

```python
# =====================================================
# ğŸ”Ÿ í‰ê°€
# =====================================================
from sklearn.metrics import (
    classification_report,
    roc_auc_score,
    average_precision_score
)

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

print("\nğŸ“Š ê¸°ë³¸ ì„±ëŠ¥")
print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
print("PR-AUC :", average_precision_score(y_test, y_prob))
```

```python
# =====================================================
# 1ï¸âƒ£1ï¸âƒ£ Threshold ì‹¤í—˜
# =====================================================
for t in [0.3, 0.4, 0.5, 0.6, 0.7]:
    print(f"\n===== Threshold = {t} =====")
    y_pred_t = (y_prob >= t).astype(int)
    print(classification_report(y_test, y_pred_t))
```
