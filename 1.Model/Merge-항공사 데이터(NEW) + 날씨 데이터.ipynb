```python
# =====================================================
# 3ï¸âƒ£ weather ì§€ì ëª… â†’ ì¶œë°œì§€ ë§¤í•‘
# =====================================================
weather_to_departure_map = {
    "ê´‘ì£¼": "ê´‘ì£¼",
    "ëŒ€êµ¬": "ëŒ€êµ¬",
    "êµ°ì‚°": "êµ°ì‚°",
    "ê¹€í¬ê³µí•­": "ê¹€í¬",
    "ê¹€í•´ì‹œ": "ê¹€í•´",
    "ë¬´ì•ˆê³µí•­": "ë¬´ì•ˆ",
    "ì–‘ì–‘ê³µí•­": "ì–‘ì–‘",
    "ì—¬ìˆ˜ê³µí•­": "ì—¬ìˆ˜",
    "ìš¸ì‚°ê³µí•­": "ìš¸ì‚°",
    "ì›ì£¼": "ì›ì£¼",
    "ì¸ì²œê³µí•­": "ì¸ì²œ",
    "ì œì£¼ê³µí•­": "ì œì£¼",
    "ì²­ì£¼": "ì²­ì£¼",
    "í¬í•­": "í¬í•­ê²½ì£¼",
    "ì§„ì£¼": "ì‚¬ì²œ"
}
```

```python
# âœ… ì§€ì ëª… â†’ ì¶œë°œì§€ ìƒì„± (í•µì‹¬)
df_weather["ì¶œë°œì§€"] = df_weather["ì§€ì ëª…"].map(weather_to_departure_map)

print("\në‚ ì”¨ ì¶œë°œì§€ ìœ ë‹ˆí¬:")
print(sorted(df_weather["ì¶œë°œì§€"].dropna().unique()))
```

```python
# =====================================================
# 4ï¸âƒ£ weather_datetime ìƒì„±
# =====================================================
df_weather["weather_datetime"] = pd.to_datetime(
    df_weather["ì¼ì‹œ"],
    errors="coerce"
)
```

```python
# =====================================================
# 1ï¸âƒ£ ì •ë ¬ (í•„ìˆ˜)
# =====================================================
df = df.sort_values("departure_datetime")
df_weather = df_weather.sort_values("weather_datetime")

# =====================================================
# 6ï¸âƒ£ merge_asof (ì¶œë°œì§€ + ì‹œê°„ ê¸°ì¤€, ê°€ì¥ ê°€ê¹Œìš´ ì •ì‹œ)
# =====================================================

df_merged = pd.merge_asof(
    df,
    df_weather,
    left_on="departure_datetime",
    right_on="weather_datetime",
    left_by="ì¶œë°œì§€",
    right_by="ì¶œë°œì§€",
    direction="nearest"      # ğŸ”¥ í•µì‹¬ ë³€ê²½
)
```

```python
# =====================================================
# 7ï¸âƒ£ ë¨¸ì§€ ê²°ê³¼ ê²€ì¦
# =====================================================
print("\nğŸ“Š ë¨¸ì§€ ê²°ê³¼ ìš”ì•½")
print("ì „ì²´ flight í–‰ ìˆ˜:", len(df_merged))
print("ë‚ ì”¨ ë¶™ì€ í–‰ ìˆ˜:", df_merged["weather_datetime"].notna().sum())
print("ë‚ ì”¨ ë¯¸ë¶€ì°© í–‰ ìˆ˜:", df_merged["weather_datetime"].isna().sum())
print("ë‚ ì”¨ ë¶€ì°© ë¹„ìœ¨:", df_merged["weather_datetime"].notna().mean())

print("\nì¶œë°œì§€ë³„ ë‚ ì”¨ ë¶€ì°©ë¥ ")
print(
    df_merged
    .groupby("ì¶œë°œì§€")["weather_datetime"]
    .apply(lambda x: x.notna().mean())
    .sort_values()
)
```

```python
# =====================================================
# âŒ ë‚ ì”¨ ë¯¸ë¶€ì°© í–‰ ì œê±°
# =====================================================
before_cnt = len(df_merged)

df_merged = df_merged[df_merged["weather_datetime"].notna()].copy()

after_cnt = len(df_merged)

print(f"ğŸ§¹ ë‚ ì”¨ ë¯¸ë¶€ì°© í–‰ ì œê±° ì™„ë£Œ")
print(f"ì œê±° ì „ í–‰ ìˆ˜: {before_cnt}")
print(f"ì œê±° í›„ í–‰ ìˆ˜: {after_cnt}")
print(f"ì œê±°ëœ í–‰ ìˆ˜: {before_cnt - after_cnt}")
```

```python
# =====================================================
# 8ï¸âƒ£ ìµœì¢… CSV ì €ì¥
# =====================================================
save_path = "new_flight_weather_merged.csv"

df_merged.to_csv(
    save_path,
    index=False,
    encoding="utf-8-sig"
)

print(f"\nâœ… CSV ì €ì¥ ì™„ë£Œ: {save_path}")
```
