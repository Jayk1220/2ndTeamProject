# 시간 정리

```python
def time_tream(time_val):
    if pd.isna(time_val) or time_val == 'nan' or time_val == '':
        return None
    
    time_str = str(time_val).replace(':','').strip()
    if time_str.isdigit():
        return time_str.zfill(4)
    
    return time_str

time_columns = ['계획시간','예상시간','출발시간','도착시간']
for col in time_columns:
    df[col] = df[col].apply(time_tream)

print(df[time_columns].head())
```

# 지연 시간

```python
import pandas as pd
import numpy as np

# --- 1. '도착' 데이터 처리 ---
target_mask_arr = df['출발/도착'] == '도착'

# .astype(str)을 붙여서 확실하게 문자열로 변환한 뒤 zfill을 적용합니다.
s_time_arr = df.loc[target_mask_arr, '계획시간'].astype(str).str.zfill(4)
a_time_arr = df.loc[target_mask_arr, '도착시간'].astype(str).str.zfill(4) 

# 결측치 문자열 처리
s_time_arr = s_time_arr.replace(['nan', 'None', '000nan'], np.nan)
a_time_arr = a_time_arr.replace(['nan', 'None', '000nan'], np.nan)

# 변환 시 다시 한번 .astype(str)을 호출하여 .str 접근자 에러를 방지합니다.
sched_min_arr = s_time_arr.astype(str).str[:2].astype(float) * 60 + s_time_arr.astype(str).str[2:].astype(float)
actual_min_arr = a_time_arr.astype(str).str[:2].astype(float) * 60 + a_time_arr.astype(str).str[2:].astype(float)

# 결과 할당
df.loc[target_mask_arr, '지연_분'] = actual_min_arr - sched_min_arr


# --- 2. '출발' 데이터 처리 (동일한 로직 적용) ---
target_mask_dep = df['출발/도착'] == '출발'

s_time_dep = df.loc[target_mask_dep, '계획시간'].astype(str).str.zfill(4)
a_time_dep = df.loc[target_mask_dep, '출발시간'].astype(str).str.zfill(4)

s_time_dep = s_time_dep.replace(['nan', 'None', '000nan'], np.nan)
a_time_dep = a_time_dep.replace(['nan', 'None', '000nan'], np.nan)

sched_min_dep = s_time_dep.astype(str).str[:2].astype(float) * 60 + s_time_dep.astype(str).str[2:].astype(float)
actual_min_dep = a_time_dep.astype(str).str[:2].astype(float) * 60 + a_time_dep.astype(str).str[2:].astype(float)

df.loc[target_mask_dep, '지연_분'] = actual_min_dep - sched_min_dep

# 확인
print(df[['출발/도착', '계획시간', '출발시간', '지연_분']].head())
```
