```python
# 회황 회항 오타 수정 및 정상운항 통합
df['상태'] = df['상태'].replace('회황', '회항')
df['상태'] = df['상태'].replace('도착', '정상운항')
df['상태'] = df['상태'].replace('출발', '정상운항')
df['상태'] = df['상태'].replace('착륙', '정상운항')
```

# 시간 정리

```python
def time_tream(time_val):
    if pd.isna(time_val) or time_val == 'nan' or time_val == '':
        return None
    
    time_str = str(time_val).replace(':','').strip()
    if time_str.isdigit():
        return time_str.zfill(4)
    
    return time_str

time_columns = ['계획시간','예상시간','출발시간']
for col in time_columns:
    df[col] = df[col].apply(time_tream)

print(df[time_columns].head())
```

# 지연 시간

```python
import pandas as pd
import numpy as np

# 1. 원본 데이터에서 결측치가 아닌 것들만 처리하기 위해 copy (권장)
# zfill(4)를 통해 505 -> 0505 형태로 보정
s_time = df['계획시간'].astype(str).str.zfill(4)
a_time = df['출발시간'].astype(str).str.zfill(4)

# 2. 'nan' 문자열로 변환된 결측치들을 진짜 NaN(None)으로 인식시키기
s_time = s_time.replace(['nan', 'None', '000nan'], np.nan)
a_time = a_time.replace(['nan', 'None', '000nan'], np.nan)

# 3. 시간과 분을 숫자로 변환 (결측치가 있으면 결과도 NaN이 됨)
sched_min = s_time.str[:2].astype(float) * 60 + s_time.str[2:].astype(float)
actual_min = a_time.str[:2].astype(float) * 60 + a_time.str[2:].astype(float)

# 4. 차이 계산
df['지연_분'] = actual_min - sched_min

# 5. 날짜 변경 보정 (값이 있는 것들에 대해서만 수행)
df.loc[df['지연_분'] < -1200, '지연_분'] += 1440
df.loc[df['지연_분'] > 1200, '지연_분'] -= 1440

# 6. 정수형으로 보고 싶다면 (결측치가 있으면 float 유지가 기본이지만, 확인용)
print(df[['계획시간', '출발시간', '지연_분']].head(10))
```
