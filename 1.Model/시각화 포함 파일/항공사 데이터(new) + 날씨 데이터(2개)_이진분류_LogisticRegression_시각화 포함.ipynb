```python
import pandas as pd
import numpy as np
```

```python
csv_path = "new_flight_weather_merged.csv"
df = pd.read_csv(csv_path)

print("âœ… CSV ë¡œë“œ ì™„ë£Œ")
print("í–‰ ìˆ˜:", len(df))
print("ì»¬ëŸ¼ ìˆ˜:", df.shape[1])
df.head()
```

```python
# =====================================================
# ğŸ” ìˆ˜ì¹˜í˜• / ë²”ì£¼í˜• ì»¬ëŸ¼ ìë™ ë¶„ë¥˜
# =====================================================

# ğŸ”¢ ìˆ˜ì¹˜í˜• ì»¬ëŸ¼
num_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()

# ğŸ· ë²”ì£¼í˜• ì»¬ëŸ¼
cat_cols = df.select_dtypes(include=["object"]).columns.tolist()

print("\nğŸ”¢ ìˆ˜ì¹˜í˜• ì»¬ëŸ¼ ìˆ˜:", len(num_cols))
print(num_cols)

print("\nğŸ· ë²”ì£¼í˜• ì»¬ëŸ¼ ìˆ˜:", len(cat_cols))
print(cat_cols)
```

```python
# =====================================================
# ğŸ”  ê²°ì¸¡ì¹˜ ì ê²€
# =====================================================

print("\nğŸ”¢ ìˆ˜ì¹˜í˜• ì»¬ëŸ¼ ê²°ì¸¡ì¹˜ í˜„í™©")
num_missing = (
    df[num_cols]
    .isna()
    .sum()
    .reset_index()
    .rename(columns={"index": "column", 0: "missing_cnt"})
)
num_missing["missing_ratio"] = num_missing["missing_cnt"] / len(df)
num_missing = num_missing[num_missing["missing_cnt"] > 0]

display(num_missing.sort_values("missing_cnt", ascending=False))


print("\nğŸ· ë²”ì£¼í˜• ì»¬ëŸ¼ ê²°ì¸¡ì¹˜ í˜„í™©")
cat_missing = (
    df[cat_cols]
    .isna()
    .sum()
    .reset_index()
    .rename(columns={"index": "column", 0: "missing_cnt"})
)
cat_missing["missing_ratio"] = cat_missing["missing_cnt"] / len(df)
cat_missing = cat_missing[cat_missing["missing_cnt"] > 0]

display(cat_missing.sort_values("missing_cnt", ascending=False))
```

```python
# =====================================================
# 5ï¸âƒ£ ë²”ì£¼í˜• / ìˆ˜ì¹˜í˜• ì»¬ëŸ¼ (ìˆ˜ì¹˜í˜• 4ê°œë§Œ ì‚¬ìš©)
# =====================================================

# ğŸ· ë²”ì£¼í˜•
cat_cols = [
    "ê³µí•­ëª…",
    "ì¶œë°œì§€",
    "ë„ì°©ì§€",
    "flight_type"
]
cat_cols = [c for c in cat_cols if c in df.columns]

# ğŸ”¢ ìˆ˜ì¹˜í˜• (ì™„ì „ ê³ ì •)
num_cols = [
    "ê¸°ì˜¨(Â°C)",
    "í’ì†_ms",
    "dep_hour",
    "dep_weekday",
    "is_weekend"
]


num_cols = [c for c in num_cols if c in df.columns]

print("\nğŸ“Š ìµœì¢… ìˆ˜ì¹˜í˜• ì»¬ëŸ¼ ìˆ˜:", len(num_cols))
print(num_cols)

print("\nğŸ· ìµœì¢… ë²”ì£¼í˜• ì»¬ëŸ¼ ìˆ˜:", len(cat_cols))
print(cat_cols)
```

```python
# =====================================================
# â± departure_datetime íƒ€ì… ë³€í™˜ (í•„ìˆ˜)
# =====================================================
df["departure_datetime"] = pd.to_datetime(df["departure_datetime"])
```

```python
# =====================================================
# 7ï¸âƒ£ Train / Test ë¶„ë¦¬ (ë‚ ì§œ ê¸°ì¤€)
# =====================================================
df = df.sort_values("departure_datetime")

split_date = df["departure_datetime"].quantile(0.8)

train_df = df[df["departure_datetime"] <= split_date]
test_df  = df[df["departure_datetime"] > split_date]

print("\nğŸ“¦ ë°ì´í„° ë¶„ë¦¬")
print("Train:", len(train_df), "Test:", len(test_df))

X_cols = num_cols + cat_cols

X_train = train_df[X_cols]
y_train = train_df["is_delay"]

X_test  = test_df[X_cols]
y_test  = test_df["is_delay"]
```

```python
# =====================================================
# ğŸ” ìˆ˜ì¹˜í˜• ë³€ìˆ˜ â†” is_delay (Spearman)
# =====================================================

corr_df = (
    train_df[num_cols + ["is_delay"]]
    .dropna()   # ìƒê´€ê³„ìˆ˜ ê³„ì‚° ê°€ëŠ¥í•œ í–‰ë§Œ ì‚¬ìš©
    .corr(method="spearman")
)

result = (
    corr_df["is_delay"]
    .drop("is_delay")
    .reset_index()
)
result.columns = ["feature", "spearman_corr"]
result["abs_corr"] = result["spearman_corr"].abs()

print("ğŸ“Š ìˆ˜ì¹˜í˜• ë³€ìˆ˜ â†” is_delay Spearman ìƒê´€ê´€ê³„")
display(result.sort_values("abs_corr", ascending=False))
```

```python
# =====================================================
# ğŸ” Bï¸âƒ£ ë²”ì£¼í˜• ë³€ìˆ˜ â†” is_delay (Chi-square)
# =====================================================
from scipy.stats import chi2_contingency

cat_corr_results = []

for col in cat_cols:
    contingency = pd.crosstab(train_df[col], train_df["is_delay"])

    # ëª¨ë“  ê°’ì´ í•œìª½ í´ë˜ìŠ¤ë§Œ ê°€ì§€ë©´ ìŠ¤í‚µ
    if contingency.shape[1] < 2:
        continue

    chi2, p, _, _ = chi2_contingency(contingency)

    cat_corr_results.append({
        "feature": col,
        "chi2": chi2,
        "p_value": p
    })

cat_corr_df = (
    pd.DataFrame(cat_corr_results)
    .sort_values("chi2", ascending=False)
)

print("\nğŸ· ë²”ì£¼í˜• ë³€ìˆ˜ Chi-square ê²°ê³¼")
display(cat_corr_df)
```

```python
# =====================================================
# 8ï¸âƒ£ ì „ì²˜ë¦¬ + ëª¨ë¸ ì •ì˜ (ê²°ì¸¡ì¹˜ í¬í•¨)
# =====================================================
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer

# ğŸ”¢ ìˆ˜ì¹˜í˜• ì „ì²˜ë¦¬
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median"))
])

# ğŸ· ë²”ì£¼í˜• ì „ì²˜ë¦¬
cat_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="constant", fill_value="UNKNOWN")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", num_pipeline, num_cols),
        ("cat", cat_pipeline, cat_cols)
    ]
)

model = Pipeline([
    ("prep", preprocessor),
    ("clf", LogisticRegression(
    max_iter=5000,
    n_jobs=-1,
    class_weight={0:1, 1:3}   # ë˜ëŠ” 5
    ))
])
```

```python
# =====================================================
# 9ï¸âƒ£ ëª¨ë¸ í•™ìŠµ
# =====================================================
model.fit(X_train, y_train)
print("\nâœ… ëª¨ë¸ í•™ìŠµ ì™„ë£Œ")
```

```python
# =====================================================
# ğŸ”Ÿ í‰ê°€
# =====================================================
from sklearn.metrics import (
    classification_report,
    roc_auc_score,
    average_precision_score
)

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

print("\nğŸ“Š ê¸°ë³¸ ì„±ëŠ¥")
print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
print("PR-AUC :", average_precision_score(y_test, y_prob))
```

```python
# =====================================================
# 1ï¸âƒ£1ï¸âƒ£ Threshold ì‹¤í—˜
# =====================================================
for t in [0.3, 0.4, 0.5, 0.6, 0.7]:
    print(f"\n===== Threshold = {t} =====")
    y_pred_t = (y_prob >= t).astype(int)
    print(classification_report(y_test, y_pred_t))
```

## ğŸ“Š ì‹œê°í™”: ì„±ëŠ¥ ìš”ì•½ í‘œ & ê·¸ë˜í”„

ì•„ë˜ ì…€ë“¤ì€ **ê¸°ì¡´ ì½”ë“œ(í•™ìŠµ/ì˜ˆì¸¡/í‰ê°€)** ì‹¤í–‰ì´ ëë‚œ ë’¤, ë§¨ ì•„ë˜ì—ì„œ ì¶”ê°€ë¡œ ì‹¤í–‰í•˜ë©´ ë©ë‹ˆë‹¤.  
(ê¸°ì¡´ ì½”ë“œëŠ” ê·¸ëŒ€ë¡œ ìœ ì§€í–ˆê³ , ì‹œê°í™”ë§Œ "ì¶”ê°€"í–ˆìŠµë‹ˆë‹¤.)


```python
# =====================================================
# ğŸ“Œ ì‹œê°í™” ì¤€ë¹„: ë³€ìˆ˜ ìë™ íƒìƒ‰ + ê¸°ë³¸ ì„±ëŠ¥ ìš”ì•½ í‘œ ìƒì„±
# -----------------------------------------------------
# âœ… ì´ ì…€ì€ "ê¸°ì¡´ ì½”ë“œ"ì—ì„œ ë§Œë“  ë³€ìˆ˜ë“¤ì„ ìµœëŒ€í•œ ìë™ìœ¼ë¡œ ì°¾ì•„ì„œ,
#    - y_true (ì •ë‹µ)
#    - y_proba (ì˜ˆì¸¡ í™•ë¥ )
#    - y_pred (ì˜ˆì¸¡ ë¼ë²¨)
#    - model (í•™ìŠµ ëª¨ë¸)
#    - X_test (í…ŒìŠ¤íŠ¸ ë…ë¦½ë³€ìˆ˜)
# ë¥¼ í™•ë³´í•œ ë’¤, í‘œë¡œ ì •ë¦¬í•©ë‹ˆë‹¤.
#
# âš ï¸ ë…¸íŠ¸ë¶ë§ˆë‹¤ ë³€ìˆ˜ëª…ì´ ì¡°ê¸ˆì”© ë‹¤ë¥¼ ìˆ˜ ìˆì–´ì„œ,
#    ì•„ë˜ ì½”ë“œê°€ ì—¬ëŸ¬ í›„ë³´ ì´ë¦„ì„ ìˆœì„œëŒ€ë¡œ íƒìƒ‰í•©ë‹ˆë‹¤.
# =====================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    average_precision_score
)

def _pick_first_existing(name_candidates):
    # ì—¬ëŸ¬ ë³€ìˆ˜ëª… í›„ë³´ ì¤‘ì—ì„œ í˜„ì¬ ë…¸íŠ¸ë¶ ì‹¤í–‰ í™˜ê²½(globals)ì— ì¡´ì¬í•˜ëŠ”
    # 'ì²« ë²ˆì§¸' ë³€ìˆ˜ë¥¼ ì°¾ì•„ (ê°’, ë³€ìˆ˜ëª…) í˜•íƒœë¡œ ë°˜í™˜
    for n in name_candidates:
        if n in globals():
            return globals()[n], n
    return None, None

def _as_1d_array(x):
    # ë¦¬ìŠ¤íŠ¸/ì‹œë¦¬ì¦ˆ/ë„˜íŒŒì´ ë“± -> 1ì°¨ì› np.arrayë¡œ ì•ˆì „í•˜ê²Œ ë³€í™˜
    if x is None:
        return None
    try:
        arr = np.array(x)
        # (N,1) ë˜ëŠ” (1,N) ê°™ì€ í˜•íƒœë©´ 1ì°¨ì›ìœ¼ë¡œ í¼ì¹¨
        return arr.reshape(-1)
    except Exception:
        return None

# -----------------------------------------------------
# 1) y_true(ì •ë‹µ) ìë™ íƒìƒ‰
# -----------------------------------------------------
y_true, y_true_name = _pick_first_existing([
    "y_test", "test_y", "y_valid", "y_val", "y_true", "Y_test", "Y_val"
])
y_true = _as_1d_array(y_true)

# -----------------------------------------------------
# 2) X_test(í…ŒìŠ¤íŠ¸ ë…ë¦½ë³€ìˆ˜) ìë™ íƒìƒ‰ (í”¼ì²˜ëª… ì¶”ì¶œìš©)
# -----------------------------------------------------
X_test, X_test_name = _pick_first_existing([
    "X_test", "test_X", "X_valid", "X_val", "X_te", "x_test"
])

# -----------------------------------------------------
# 3) model(í•™ìŠµ ëª¨ë¸) ìë™ íƒìƒ‰
# -----------------------------------------------------
# âœ… ë¹„êµ ë…¸íŠ¸ë¶(ì—¬ëŸ¬ ëª¨ë¸)ë„ ìˆì„ ìˆ˜ ìˆì–´ì„œ, í”í•œ ì´ë¦„ë“¤ì„ ë„“ê²Œ íƒìƒ‰í•©ë‹ˆë‹¤.
model, model_name = _pick_first_existing([
    "model", "clf", "classifier",
    "lgb_model", "lgbm_model", "lgbm",
    "xgb_model", "xgb", "xgb_clf",
    "best_model", "final_model"
])

# -----------------------------------------------------
# 4) y_proba(ì˜ˆì¸¡ í™•ë¥ ) ìë™ íƒìƒ‰
# -----------------------------------------------------
# âœ… ì´ë¯¸ ë§Œë“¤ì–´ë‘” í™•ë¥  ë³€ìˆ˜ê°€ ìˆìœ¼ë©´ ê·¸ê±¸ ìš°ì„  ì‚¬ìš©í•©ë‹ˆë‹¤.
y_proba, y_proba_name = _pick_first_existing([
    "y_proba", "y_pred_proba", "pred_proba", "proba",
    "y_score", "scores", "prob", "p_pred"
])

# âœ… í™•ë¥  ë³€ìˆ˜ê°€ ì—†ìœ¼ë©´, model + X_testë¡œ ì§ì ‘ ê³„ì‚°ì„ ì‹œë„í•©ë‹ˆë‹¤.
if y_proba is None and model is not None and X_test is not None:
    try:
        # sklearn / xgboost / lightgbm ëŒ€ë¶€ë¶„ì€ predict_proba ì§€ì›
        if hasattr(model, "predict_proba"):
            proba = model.predict_proba(X_test)
            # ë³´í†µ ì´ì§„ë¶„ë¥˜ëŠ” (N, 2) -> positive class í™•ë¥ ì€ [:, 1]
            if isinstance(proba, np.ndarray) and proba.ndim == 2 and proba.shape[1] >= 2:
                y_proba = proba[:, 1]
                y_proba_name = f"{model_name}.predict_proba({X_test_name})[:,1]"
            else:
                # í˜¹ì‹œ (N,)ìœ¼ë¡œ ë°”ë¡œ ë‚˜ì˜¤ëŠ” ê²½ìš°
                y_proba = proba
                y_proba_name = f"{model_name}.predict_proba({X_test_name})"
        # decision_functionë§Œ ìˆëŠ” ëª¨ë¸ë„ ì¡´ì¬í•  ìˆ˜ ìˆìŒ (SVM ë“±)
        elif hasattr(model, "decision_function"):
            score = model.decision_function(X_test)
            y_proba = score
            y_proba_name = f"{model_name}.decision_function({X_test_name})"
    except Exception as e:
        print("âš ï¸ modelë¡œë¶€í„° ì˜ˆì¸¡ í™•ë¥ (y_proba) ê³„ì‚° ì‹¤íŒ¨:", repr(e))

y_proba = _as_1d_array(y_proba)

# -----------------------------------------------------
# 5) y_pred(ì˜ˆì¸¡ ë¼ë²¨) ìë™ íƒìƒ‰
# -----------------------------------------------------
y_pred, y_pred_name = _pick_first_existing([
    "y_pred", "pred", "y_hat", "y_pred_label"
])

# âœ… y_predê°€ ì—†ìœ¼ë©´, y_probaë¡œ threshold=0.5 ê¸°ì¤€ ë¼ë²¨ ìƒì„±
if y_pred is None and y_proba is not None:
    y_pred = (y_proba >= 0.5).astype(int)
    y_pred_name = "derived_from_y_proba(threshold=0.5)"

y_pred = _as_1d_array(y_pred)

# -----------------------------------------------------
# 6) í™•ë³´í•œ ë³€ìˆ˜ë“¤ ìš”ì•½ ì¶œë ¥
# -----------------------------------------------------
print("âœ… [ìë™ íƒìƒ‰ ê²°ê³¼]")
print(f" - y_true  : {y_true_name}")
print(f" - y_proba : {y_proba_name}")
print(f" - y_pred  : {y_pred_name}")
print(f" - model   : {model_name}")
print(f" - X_test  : {X_test_name}")

# -----------------------------------------------------
# 7) ê¸°ë³¸ ì„±ëŠ¥ ìš”ì•½ í‘œ(ì´ì§„ë¶„ë¥˜ ê¸°ì¤€)
# -----------------------------------------------------
# âš ï¸ y_true ë˜ëŠ” y_predê°€ ì—†ìœ¼ë©´ í‘œë¥¼ ë§Œë“¤ ìˆ˜ ì—†ìœ¼ë‹ˆ ì•ˆë‚´ í›„ ì¢…ë£Œí•©ë‹ˆë‹¤.
if y_true is None or y_pred is None:
    raise ValueError("y_true ë˜ëŠ” y_predë¥¼ ì°¾ì§€ ëª»í–ˆìŠµë‹ˆë‹¤. (ê¸°ì¡´ ì½”ë“œì—ì„œ test ì •ë‹µ/ì˜ˆì¸¡ ë³€ìˆ˜ë¥¼ ë§Œë“  ë’¤ ë‹¤ì‹œ ì‹¤í–‰í•˜ì„¸ìš”.)")

# âœ… classification_reportë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜ (í‘œë¡œ ë³´ê¸° ì¢‹ê²Œ)
report_dict = classification_report(y_true, y_pred, output_dict=True, zero_division=0)
report_df = pd.DataFrame(report_dict).T

# âœ… ì „ì²´ ìŠ¤ì¹¼ë¼ ì§€í‘œ(ìˆì„ ë•Œë§Œ ê³„ì‚°)
metrics = {
    "accuracy": accuracy_score(y_true, y_pred),
    "precision(binary)": precision_score(y_true, y_pred, zero_division=0),
    "recall(binary)": recall_score(y_true, y_pred, zero_division=0),
    "f1(binary)": f1_score(y_true, y_pred, zero_division=0),
}

# âœ… í™•ë¥ ì´ ìˆìœ¼ë©´ ROC-AUC / PR-AUCë„ ê³„ì‚°
if y_proba is not None:
    try:
        metrics["roc_auc"] = roc_auc_score(y_true, y_proba)
    except Exception:
        metrics["roc_auc"] = np.nan
    try:
        metrics["pr_auc(AP)"] = average_precision_score(y_true, y_proba)
    except Exception:
        metrics["pr_auc(AP)"] = np.nan
else:
    metrics["roc_auc"] = np.nan
    metrics["pr_auc(AP)"] = np.nan

metrics_df = pd.DataFrame([metrics])

print("\nâœ… [ì„±ëŠ¥ ìš”ì•½(ìŠ¤ì¹¼ë¼)]")
display(metrics_df)

print("\nâœ… [Classification Report (í‘œ)]")
display(report_df)

# -----------------------------------------------------
# 8) Confusion Matrix(í˜¼ë™í–‰ë ¬) í‘œ
# -----------------------------------------------------
cm = confusion_matrix(y_true, y_pred)
cm_df = pd.DataFrame(cm, index=["Actual 0", "Actual 1"], columns=["Pred 0", "Pred 1"])

print("\nâœ… [Confusion Matrix (í‘œ)]")
display(cm_df)

# -----------------------------------------------------
# 9) Threshold(ì„ê³„ê°’) ìŠ¤ìœ• í…Œì´ë¸”
# -----------------------------------------------------
# âœ… í™•ë¥ (y_proba)ì´ ìˆì„ ë•Œë§Œ ì˜ë¯¸ê°€ ìˆìœ¼ë¯€ë¡œ, ì—†ìœ¼ë©´ ìŠ¤í‚µí•©ë‹ˆë‹¤.
if y_proba is not None:
    thresholds = np.linspace(0.05, 0.95, 19)
    rows = []
    for t in thresholds:
        yp = (y_proba >= t).astype(int)
        rows.append({
            "threshold": float(np.round(t, 2)),
            "accuracy": accuracy_score(y_true, yp),
            "precision": precision_score(y_true, yp, zero_division=0),
            "recall": recall_score(y_true, yp, zero_division=0),
            "f1": f1_score(y_true, yp, zero_division=0),
        })
    thr_df = pd.DataFrame(rows).sort_values("threshold")
    print("\nâœ… [Threshold ìŠ¤ìœ• ì„±ëŠ¥ í‘œ] (í™•ë¥  ê¸°ë°˜)")
    display(thr_df)
else:
    thr_df = None
    print("\nâ„¹ï¸ y_proba(ì˜ˆì¸¡ í™•ë¥ )ê°€ ì—†ì–´ threshold ìŠ¤ìœ• í‘œëŠ” ìƒëµí•©ë‹ˆë‹¤.")
```

```python
# =====================================================
# ğŸ“ˆ ì‹œê°í™”: í˜¼ë™í–‰ë ¬ / ROC / PR / í™•ë¥ ë¶„í¬ / ì„ê³„ê°’ ê³¡ì„  / ì¤‘ìš” í”¼ì²˜
# -----------------------------------------------------
# âœ… ì´ ì…€ì€ ìœ„ ì…€ì—ì„œ ë§Œë“ (ë˜ëŠ” ì°¾ì€) ë³€ìˆ˜(y_true, y_pred, y_proba, model, X_test)ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.
# âœ… ê·¸ë˜í”„ëŠ” matplotlib ê¸°ë³¸ ìŠ¤íƒ€ì¼(ìƒ‰ ì§€ì • ì—†ìŒ)ë¡œ ê·¸ë¦½ë‹ˆë‹¤.
# =====================================================

from sklearn.metrics import roc_curve, precision_recall_curve

# -----------------------------------------------------
# 1) í˜¼ë™í–‰ë ¬(Confusion Matrix) ì‹œê°í™”
# -----------------------------------------------------
plt.figure(figsize=(5, 4))
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.colorbar()

tick_marks = np.arange(2)
plt.xticks(tick_marks, ["Pred 0", "Pred 1"])
plt.yticks(tick_marks, ["Actual 0", "Actual 1"])

# âœ… ì…€ ì•ˆì— ìˆ«ì(ê±´ìˆ˜) í‘œì‹œ
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, str(cm[i, j]), ha="center", va="center")

plt.tight_layout()
plt.show()

# -----------------------------------------------------
# 2) ROC Curve (í™•ë¥ ì´ ìˆì„ ë•Œë§Œ)
# -----------------------------------------------------
if y_proba is not None:
    fpr, tpr, _ = roc_curve(y_true, y_proba)
    plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr, label="ROC")
    plt.plot([0, 1], [0, 1], linestyle="--", label="Random")
    plt.title("ROC Curve")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.tight_layout()
    plt.show()

# -----------------------------------------------------
# 3) Precision-Recall Curve (í™•ë¥ ì´ ìˆì„ ë•Œë§Œ)
# -----------------------------------------------------
if y_proba is not None:
    precision, recall, _ = precision_recall_curve(y_true, y_proba)
    plt.figure(figsize=(6, 4))
    plt.plot(recall, precision, label="PR")
    plt.title("Precision-Recall Curve")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.legend()
    plt.tight_layout()
    plt.show()

# -----------------------------------------------------
# 4) ì˜ˆì¸¡ í™•ë¥  ë¶„í¬(í´ë˜ìŠ¤ë³„) (í™•ë¥ ì´ ìˆì„ ë•Œë§Œ)
# -----------------------------------------------------
# âœ… í´ë˜ìŠ¤ 0ê³¼ 1ì˜ ì˜ˆì¸¡ í™•ë¥  ë¶„í¬ë¥¼ ë”°ë¡œ ê·¸ë¦¬ë©´,
#    ëª¨ë¸ì´ ì–´ëŠ ì •ë„ë¡œ ë‘ í´ë˜ìŠ¤ë¥¼ "ë¶„ë¦¬"í•˜ëŠ”ì§€ ì§ê´€ì ìœ¼ë¡œ í™•ì¸ ê°€ëŠ¥í•©ë‹ˆë‹¤.
if y_proba is not None:
    y_proba_0 = y_proba[y_true == 0]
    y_proba_1 = y_proba[y_true == 1]

    plt.figure(figsize=(6, 4))
    plt.hist(y_proba_0, bins=50, alpha=0.6, label="Actual 0")
    plt.hist(y_proba_1, bins=50, alpha=0.6, label="Actual 1")
    plt.title("Predicted Probability Distribution")
    plt.xlabel("Predicted probability (positive class)")
    plt.ylabel("Count")
    plt.legend()
    plt.tight_layout()
    plt.show()

# -----------------------------------------------------
# 5) Threshold(ì„ê³„ê°’) ë³€í™”ì— ë”°ë¥¸ Precision/Recall/F1 ê³¡ì„  (í™•ë¥ ì´ ìˆì„ ë•Œë§Œ)
# -----------------------------------------------------
if thr_df is not None:
    plt.figure(figsize=(7, 4))
    plt.plot(thr_df["threshold"], thr_df["precision"], label="precision")
    plt.plot(thr_df["threshold"], thr_df["recall"], label="recall")
    plt.plot(thr_df["threshold"], thr_df["f1"], label="f1")
    plt.title("Precision / Recall / F1 vs Threshold")
    plt.xlabel("Threshold")
    plt.ylabel("Score")
    plt.legend()
    plt.tight_layout()
    plt.show()

# -----------------------------------------------------
# 6) ì¤‘ìš” í”¼ì²˜(Feature Importance / Coef) Top 20
# -----------------------------------------------------
# âœ… íŠ¸ë¦¬ ê¸°ë°˜ ëª¨ë¸(lightgbm/xgboost ë“±): feature_importances_
# âœ… ì„ í˜• ëª¨ë¸(logistic regression ë“±): coef_
#
# âš ï¸ í”¼ì²˜ëª…ì´ í•„ìš”í•©ë‹ˆë‹¤. ë³´í†µ X_testê°€ DataFrameì´ë©´ columnsë¡œ ê°€ì ¸ì˜µë‹ˆë‹¤.
#    (numpy arrayì´ë©´, í”¼ì²˜ëª…ì„ ëª¨ë¥¼ ìˆ˜ ìˆì–´ indexë¡œ í‘œì‹œí•©ë‹ˆë‹¤.)
# -----------------------------------------------------
feature_names = None
if X_test is not None:
    try:
        if hasattr(X_test, "columns"):
            feature_names = list(X_test.columns)
        else:
            # numpy arrayì¸ ê²½ìš°: í”¼ì²˜ëª… ì—†ìŒ -> ë²ˆí˜¸ë¡œ ëŒ€ì²´
            feature_names = [f"f{i}" for i in range(X_test.shape[1])]
    except Exception:
        feature_names = None

def _plot_top_features(values, names, title, top_n=20):
    # ì¤‘ìš”ë„/ê³„ìˆ˜ ë°°ì—´(values)ì™€ í”¼ì²˜ëª…(names)ìœ¼ë¡œ Top-N ë§‰ëŒ€ê·¸ë˜í”„ë¥¼ ê·¸ë¦½ë‹ˆë‹¤.
    # - values: (num_features,) í˜•íƒœ
    # - names : ê¸¸ì´ num_features
    s = pd.Series(values, index=names)
    # ì ˆëŒ€ê°’ ê¸°ì¤€ Top-N (ì–‘/ìŒ ëª¨ë‘ ì¤‘ìš”í•  ìˆ˜ ìˆì–´ì„œ)
    top = s.reindex(s.abs().sort_values(ascending=False).index).head(top_n)
    top = top.iloc[::-1]  # ë³´ê¸° ì¢‹ê²Œ ì—­ìˆœ
    plt.figure(figsize=(8, 6))
    plt.barh(top.index, top.values)
    plt.title(title)
    plt.xlabel("Importance / Coefficient")
    plt.tight_layout()
    plt.show()

# âœ… modelì´ ì¡´ì¬í•˜ê³ , feature_namesê°€ ìˆì„ ë•Œë§Œ ì‹œê°í™” ì‹œë„
if model is not None and feature_names is not None:
    try:
        # 6-A) íŠ¸ë¦¬ ê¸°ë°˜ importance
        if hasattr(model, "feature_importances_"):
            imp = np.array(model.feature_importances_).reshape(-1)
            if len(imp) == len(feature_names):
                _plot_top_features(imp, feature_names, "Top Features (feature_importances_)")
            else:
                print("âš ï¸ feature_importances_ ê¸¸ì´ì™€ feature_names ê¸¸ì´ê°€ ë‹¬ë¼ importance ì‹œê°í™”ë¥¼ ìƒëµí•©ë‹ˆë‹¤.")
        # 6-B) ì„ í˜• ëª¨ë¸ coef
        elif hasattr(model, "coef_"):
            coef = np.array(model.coef_).reshape(-1)
            if len(coef) == len(feature_names):
                _plot_top_features(coef, feature_names, "Top Features (coef_)")
            else:
                print("âš ï¸ coef_ ê¸¸ì´ì™€ feature_names ê¸¸ì´ê°€ ë‹¬ë¼ coef ì‹œê°í™”ë¥¼ ìƒëµí•©ë‹ˆë‹¤.")
        else:
            print("â„¹ï¸ ì´ ëª¨ë¸ì€ feature_importances_ ë˜ëŠ” coef_ê°€ ì—†ì–´ ì¤‘ìš” í”¼ì²˜ ì‹œê°í™”ë¥¼ ìƒëµí•©ë‹ˆë‹¤.")
    except Exception as e:
        print("âš ï¸ ì¤‘ìš” í”¼ì²˜ ì‹œê°í™” ì¤‘ ì˜¤ë¥˜:", repr(e))
else:
    print("â„¹ï¸ model ë˜ëŠ” feature_namesë¥¼ í™•ë³´í•˜ì§€ ëª»í•´ ì¤‘ìš” í”¼ì²˜ ì‹œê°í™”ë¥¼ ìƒëµí•©ë‹ˆë‹¤.")

# -----------------------------------------------------
# 7) (ì˜µì…˜) ì—¬ëŸ¬ ëª¨ë¸ ë¹„êµê°€ ê°€ëŠ¥í•œ ê²½ìš°: models(dict)ê°€ ìˆìœ¼ë©´ ì„±ëŠ¥í‘œ ìƒì„±
# -----------------------------------------------------
# âœ… ë¹„êµ ë…¸íŠ¸ë¶ì—ì„œ models = {"lgb":..., "xgb":..., ...} í˜•íƒœë¡œ ê°€ì§€ê³  ìˆëŠ” ê²½ìš°ê°€ ìˆì–´ì„œ
#    ìˆìœ¼ë©´ ìë™ìœ¼ë¡œ ëŒë ¤ì„œ í‘œë¥¼ ë§Œë“¤ì–´ ì¤ë‹ˆë‹¤.
# -----------------------------------------------------
models_dict = globals().get("models", None)

if isinstance(models_dict, dict) and X_test is not None and y_true is not None:
    rows = []
    for k, m in models_dict.items():
        try:
            # ì˜ˆì¸¡ í™•ë¥  ê³„ì‚°
            if hasattr(m, "predict_proba"):
                p = m.predict_proba(X_test)
                if isinstance(p, np.ndarray) and p.ndim == 2 and p.shape[1] >= 2:
                    p = p[:, 1]
            elif hasattr(m, "decision_function"):
                p = m.decision_function(X_test)
            else:
                p = None

            # ì˜ˆì¸¡ ë¼ë²¨(ê¸°ë³¸ threshold=0.5)
            if p is not None:
                yp = (np.array(p).reshape(-1) >= 0.5).astype(int)
                roc = roc_auc_score(y_true, p)
                pr = average_precision_score(y_true, p)
            else:
                yp = m.predict(X_test)
                roc = np.nan
                pr = np.nan

            rows.append({
                "model_name": k,
                "accuracy": accuracy_score(y_true, yp),
                "precision": precision_score(y_true, yp, zero_division=0),
                "recall": recall_score(y_true, yp, zero_division=0),
                "f1": f1_score(y_true, yp, zero_division=0),
                "roc_auc": roc,
                "pr_auc(AP)": pr
            })
        except Exception as e:
            rows.append({"model_name": k, "error": repr(e)})

    compare_df = pd.DataFrame(rows)
    print("\nâœ… [ì—¬ëŸ¬ ëª¨ë¸ ë¹„êµ ì„±ëŠ¥í‘œ]")
    display(compare_df)
```
