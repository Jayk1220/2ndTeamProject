```python
# =========================
# ğŸ›¬ ë„ì°©ì§€ â†’ ê³µí•­ ì½”ë“œ ë§¤í•‘
# =========================

def map_arrival_code(x):
    """
    ë„ì°©ì§€(í•œê¸€/ì˜ë¬¸) â†’ IATA ê³µí•­ ì½”ë“œ
    - AIRPORT_DATAì— ì—†ê±°ë‚˜
    - í›ˆë ¨ê³µì—­ / ê³µì—­ / NaN ì¸ ê²½ìš° â†’ None
    """
    if pd.isna(x):
        return None

    key = str(x).strip()
    return AIRPORT_DATA.get(key)


# ë§¤í•‘ ì ìš©
df_filtered["arrival_code"] = df_filtered["ë„ì°©ì§€"].apply(map_arrival_code)

print("ğŸ›¬ ë„ì°©ì§€ ì½”ë“œ ë§¤í•‘ ì™„ë£Œ")
print(df_filtered[["ë„ì°©ì§€", "arrival_code"]].drop_duplicates().head(20))
```

```python
# =========================
# ğŸ” ë§¤í•‘ ê²°ê³¼ ê²€ì¦
# =========================

total_rows = len(df_filtered)
mapped_rows = df_filtered["arrival_code"].notna().sum()

print(f"ì „ì²´ í–‰ ìˆ˜        : {total_rows}")
print(f"ë§¤í•‘ ì„±ê³µ í–‰ ìˆ˜  : {mapped_rows}")
print(f"ë§¤í•‘ ì„±ê³µë¥       : {mapped_rows / total_rows:.2%}")

print("\nâŒ ë§¤í•‘ ì‹¤íŒ¨ ë„ì°©ì§€ ëª©ë¡")
display(
    df_filtered[df_filtered["arrival_code"].isna()]["ë„ì°©ì§€"]
    .value_counts()
)
```

```python
# =========================
# ğŸš® ë„ì°©ì§€ ì´ìƒê°’ ì œê±° (ê³µì—­ + NaN)
# =========================

df_filtered = df_filtered[
    df_filtered["arrival_code"].notna() &
    (~df_filtered["ë„ì°©ì§€"].isin(["ê³µì—­", "í›ˆë ¨ê³µì—­", "í›ˆë ¨ê³µì—­2"]))
].reset_index(drop=True)

print("ğŸš® ì œê±° í›„ í–‰ ìˆ˜:", len(df_filtered))
```

```python
print("NaN ë‚¨ì•„ìˆëŠ”ì§€:", df_filtered["arrival_code"].isna().sum())
print("ê³µì—­ ë‚¨ì•„ìˆëŠ”ì§€:")
print(df_filtered["ë„ì°©ì§€"].value_counts().head())
```

```python
# =========================
# ğŸ•’ ì¶œë°œ datetime ìƒì„±
# =========================
df_filtered["ì¶œë°œì‹œê°„_clean"] = (
    df_filtered["ì¶œë°œì‹œê°„"].astype(int).astype(str).str.zfill(4)
)

df_filtered["departure_datetime"] = pd.to_datetime(
    df_filtered["ì¼ì"].astype(str) + df_filtered["ì¶œë°œì‹œê°„_clean"],
    format="%Y%m%d%H%M",
    errors="coerce"
)

# =========================
# â± ì‹œ / ë¶„ íŒŒìƒ ë³€ìˆ˜
# =========================
df_filtered["dep_hour"] = df_filtered["departure_datetime"].dt.hour
df_filtered["dep_min"]  = df_filtered["departure_datetime"].dt.minute
df_filtered["dep_weekday"] = df_filtered["departure_datetime"].dt.weekday
df_filtered["is_weekend"] = df_filtered["dep_weekday"].isin([5, 6]).astype(int)
```

```python
# =========================
# 7ï¸âƒ£ ì§€ì—° íƒ€ê²Ÿ ë³€ìˆ˜
# =========================
def make_delay_label(x):
    if pd.isna(x):
        return 0
    return 1 if any(k in x for k in ["ì§€ì—°", "ì·¨ì†Œ", "íšŒí•­"]) else 0

df_filtered["is_delay"] = df_filtered["ìƒíƒœ"].apply(make_delay_label)

print("â± ì§€ì—° ë¶„í¬")
print(df_filtered["is_delay"].value_counts())
```

```python
# =========================
# âœˆï¸ êµ­ë‚´ / êµ­ì œ (ê³µí•­ ì½”ë“œ ê¸°ì¤€)
# =========================
DOMESTIC_CODES = {
    "ICN","GMP","CJU","PUS","TAE","KWJ","RSU","USN",
    "MWX","HIN","KPO","WJU","KUV","CJJ","YNY"
}

df_filtered["flight_type"] = df_filtered["arrival_code"].apply(
    lambda x: "êµ­ë‚´" if pd.notna(x) and x in DOMESTIC_CODES else "êµ­ì œ"
)

print("âœˆï¸ êµ­ë‚´/êµ­ì œ ë¶„í¬")
print(df_filtered["flight_type"].value_counts())
```
