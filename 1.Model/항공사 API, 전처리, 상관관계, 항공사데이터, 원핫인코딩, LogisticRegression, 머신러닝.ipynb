```python
#타겟 변수(is_delay) 생성
def make_delay_label(x):
    if pd.isna(x):
        return 0
    if any(k in x for k in ["지연", "결항", "회항"]):
        return 1
    return 0

df["is_delay"] = df["RMK_KOR"].apply(make_delay_label)
```

```python
# 날짜 파생
df["flight_date"] = pd.to_datetime(df["FLIGHT_DATE"], format="%Y%m%d")
df["weekday"] = df["flight_date"].dt.weekday
df["month"] = df["flight_date"].dt.month

# 시간 파생
df["STD"] = df["STD"].astype(str)
df["hour"] = df["STD"].str.zfill(4).str[:2].astype(int)
```

```python
#train / test 분리 (날짜 기준)

# 날짜 기준으로 정렬 (가장 중요)
df = df.sort_values("flight_date")

# 상위 80% 날짜를 학습용, 나머지 20%를 테스트용으로 사용
split_date = df["flight_date"].quantile(0.8)

# 학습 데이터 (과거)
train_df = df[df["flight_date"] <= split_date]

# 테스트 데이터 (최근)
test_df = df[df["flight_date"] > split_date]
```

```python
# 수치형 변수 ↔ 종속변수 상관관계 (Point-Biserial)

from scipy.stats import pointbiserialr

num_cols = ["weekday", "month", "hour"]

corr_results = []

for col in num_cols:
    corr, pval = pointbiserialr(train_df[col], train_df["is_delay"])
    corr_results.append({
        "feature": col,
        "corr": corr,
        "abs_corr": abs(corr),
        "p_value": pval
    })

corr_df = pd.DataFrame(corr_results).sort_values("abs_corr", ascending=False)
corr_df
```

```python
#범주형 변수 ↔ 종속변수 (Chi-square 검정)

from scipy.stats import chi2_contingency

cat_cols = [
    "AIRLINE_KOREAN",
    "BOARDING_KOR",
    "ARRIVED_KOR",
    "LINE",
    "IO"
]

chi_results = []

for col in cat_cols:
    contingency = pd.crosstab(train_df[col], train_df["is_delay"])
    chi2, p, _, _ = chi2_contingency(contingency)

    chi_results.append({
        "feature": col,
        "chi2": chi2,
        "p_value": p
    })

chi_df = pd.DataFrame(chi_results).sort_values("chi2", ascending=False)
chi_df
```

```python
# 독립변수(X) / 종속변수(y) 분리

X_columns = [
    "weekday", "month", "hour",
    "AIRLINE_KOREAN", "BOARDING_KOR", "ARRIVED_KOR",
    "LINE", "IO"
]

X_train = train_df[X_columns]
y_train = train_df["is_delay"]

X_test = test_df[X_columns]
y_test = test_df["is_delay"]
```

```python
# 소수 클래스(1)에 더 큰 가중치 부여 (클래스 불균형 대응)
class_weight = {0: 1, 1: 70}
```

```python
#원핫인코딩
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# 범주형 / 수치형 분리
cat_cols = [
    "AIRLINE_KOREAN",
    "BOARDING_KOR",
    "ARRIVED_KOR",
    "LINE",
    "IO"
]

num_cols = ["weekday", "month", "hour"]

# 범주형 → 원핫 인코딩
# 수치형 → 그대로 사용
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
        ("num", "passthrough", num_cols)
    ]
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

#모델 정의 & 학습
model = Pipeline([
    ("prep", preprocessor),  # 전처리
    ("clf", LogisticRegression(
        max_iter=5000,
        class_weight=class_weight
    ))
])
```

```python
# 학습 실행
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

# 최종 분류 결과
y_pred = model.predict(X_test)

# 예측 확률 (문제일 확률)
y_prob = model.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

```python
from sklearn.metrics import average_precision_score

pr_auc = average_precision_score(y_test, y_prob)
print("PR-AUC:", pr_auc)
```

```python
# Threshold 실험
for t in [0.3, 0.4, 0.5, 0.6, 0.7, 0.8]:
    y_pred_t = (y_prob >= t).astype(int)
    print(f"\n===== Threshold = {t} =====")
    print(classification_report(y_test, y_pred_t))
```
